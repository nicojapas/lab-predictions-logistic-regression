1. Create a query or queries to extract the information you think may be relevant for building the prediction model. It should include some film features and some rental features.

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import math
import pymysql
#from sqlalchemy import create_engine
import getpass  # To get the password without showing the input
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/bank'
```

```python
query = 'SELECT film.rental_rate, r.rental_id, cat.name as "film_category"\
FROM sakila.rental AS r \
JOIN sakila.inventory AS inv \
USING (inventory_id) \
JOIN sakila.film AS film \
USING (film_id) \
JOIN sakila.film_category AS ficat \
USING (film_id) \
JOIN sakila.category AS cat \
USING (category_id)'
```

2. Read the data into a Pandas dataframe.

```python
data = pd.read_sql_query(query,connection_string)
data.head()
```

```python
data.shape
```

3. Analyze extracted features and transform them. You may need to encode some categorical variables, or scale numerical variables.

```python
data['rental_rate'] = data['rental_rate'].astype(object)
```

```python
data.dtypes
```

4. Create a query to get the list of films and a boolean indicating if it was rented last month. This would be our target variable.

```python
query = 'select films.title, rentals.rental_id, \
        year(rentals.rental_date)=2006 and month(rentals.rental_date)=2 as "rented_last_month", \
        rentals.rental_date \
        from sakila.film as films \
        join sakila.inventory as inv \
        using (film_id) \
        join sakila.rental as rentals \
        using (inventory_id)'
```

```python
data_new = pd.read_sql_query(query,connection_string)
data_new.head()
```

```python
data_new = data_new.drop('rental_date', axis=1)
# Convert rented_last_month to boolean type
data_new['rented_last_month'] = data_new['rented_last_month'].astype(bool)
data_new.head()
```

```python
data = pd.merge(data, data_new, on='rental_id' )
data.head()
```

```python
data = data.drop('rental_id', axis=1)
```

```python
X_cat = data.select_dtypes(include=object)
X_cat.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# creating instance of labelencoder
labelencoder = LabelEncoder()# Assigning numerical values and storing in another column

X_cat_encoded = pd.DataFrame()
X_cat_encoded['title_encoded'] = labelencoder.fit_transform(X_cat['title'])
X_cat_encoded
```

```python
X_cat
```

```python
X_cat_encoded
```

```python
# One-Hot-Encoder using get_dummies
feature_encoded = pd.get_dummies(X_cat[['rental_rate', 'film_category']])
X_cat_encoded = X_cat_encoded.join(feature_encoded)
X_cat_encoded
```

```python
Y = data['rented_last_month']
Y.head()
```

```python
X = X_cat_encoded
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3)
```

5. Create a logistic regression model to predict this variable from the cleaned data.

```python
# Logistic Regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

6. Evaluate the results.

```python
y_predicted = model.predict(X_test)
```

```python
model.score(X_test, y_predicted)
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_test, y_predicted)
y_predicted = model.predict(X_test)

plt.figure(figsize=(8,8))
sns.heatmap(cm, square=True, annot=True,
           cbar_kws={"orientation": "vertical", 'fraction': 0.03, 'label': 'Correlation'})
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
